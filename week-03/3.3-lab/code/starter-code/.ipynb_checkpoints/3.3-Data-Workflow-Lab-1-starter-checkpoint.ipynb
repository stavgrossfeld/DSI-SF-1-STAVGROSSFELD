```python
# Remove $ from certain columns
for name in df.columns.values:
   # print df[name][0]
    if '$' in str(df[name][0]):
        df[name] = df[name].apply(lambda x: x.split('$')[1]).apply(lambda x: float(x))
    else: continue
df[0:10]
df.info()
```

```python
# Determine which stores were open all of 2015
# Find the first and last sales date.
mask_2015 = pd.DataFrame(sales[sales.Date.dt.year == 2015])


print mask_2015.Date.dt.year.value_counts()

open_january = mask_2015[mask_2015.Date.dt.month == 1]

open_december = mask_2015[mask_2015.Date.dt.month == 12]


full_year_stores = pd.Series(open_december['Store Number']).append(pd.Series(open_january['Store Number']))


len(set(full_year_stores))
```

```python


sales_agg = sales[sales['Store Number'].isin(set(full_year_stores))]
```
