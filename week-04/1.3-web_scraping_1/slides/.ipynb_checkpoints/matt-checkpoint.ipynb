```python
import requests

response = requests.get("http://shoptalk.com/attending-companies.php")
HTML = response.text  
HTML[0:1000]        

import pandas as pd
```

```python
from scrapy.selector import Selector
from scrapy.http import HtmlResponse



titles = Selector(text=HTML).xpath('//a/text()').extract()
titles[0:10] # the first 5 titles
```

```python
matt = []

for name in titles:
    name = name.rstrip()
    name = name.lstrip()
    name.split(',')
    name = name.replace(u'\xe3','')
    name = name.replace(u'\xe9','')
    matt.append(name)
    

    
```

```python
len(matt)
```

```python
str(matt[0])
```

```python
new = []
count = 0
for i in matt:
    count += 1
    print count
    if type(str(i)) == str:
        print str(i)
        new.append(str(i))
    else: 
        continue
```

```python
matt_data = pd.Series(new)
```
