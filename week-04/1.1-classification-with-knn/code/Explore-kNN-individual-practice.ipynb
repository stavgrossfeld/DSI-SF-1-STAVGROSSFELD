## 1. Load data and packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
import patsy
```

```python
affair = pd.read_csv('../assets/datasets/Fair.csv')
affair.head()
```

## 2. Pick predictors and target of interest

```python
affair['affair_binary'] = affair.nbaffairs.apply(lambda x: 0 if x == 0 else 1)
```

```python
formula = "affair_binary ~ C(sex) + C(child) + education + age + ym + religious + C(occupation) - 1"
Y, X = patsy.dmatrices(formula, data=affair)
```

```python
Y = np.ravel(Y)
print Y[0:10]
print X[0:5]
```

## 3. Do exploratory data analysis with metrics (correlation, etc.) and plotting

```python
affair.corr()
```

```python
sns.pairplot(affair)
```

## 4. Make X and Y cross-validation folds

BONUS: Use StratifiedKFold

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.33)
```

## 5. Create the kNN classifier from sklearn

```python
model = KNeighborsClassifier(n_neighbors = 3)
model.fit(X,Y)
print model.score(X,Y)
```

```python
model.fit(X_train, Y_train)
model.score(X_test, Y_test)
```

## 6. Cross-validate accuracy

Try out:
